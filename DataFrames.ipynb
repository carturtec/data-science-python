```python
import pandas as pd

# randn gera dados aleat√≥rios para uma tabela
from numpy.random import randn
```

```python
#criando uma tabela com 5 linhas e 4 colulnas | split faz a quebra em uma lista de strings
df = pd.DataFrame(randn(5, 4), index=[1, 2, 3, 4, 5], columns="A B C D".split())
```

```python
#cria uma nova coluna e preenche com a soma dos dados de duas colunas. No caso a coluna A+B
df["F"] = df['A'] + df['B']
```

```python
#cria um c√≥pia da tabela com a coluna F deletada, por√©m conserva a tabela original. Axis=1 coluna e axis=0 linha 
df.drop('F', axis=1)
```

```python
#ao excutar a tabela novamente, veja que a tabela original foi conservada e a coluna F ainda existe
df
```

```python
#podemos salvar a c√≥pia sem a coluna F em outra vari√°vel e manter a original
df2 = df.drop('F', axis=1)
```
