```python
import pandas as pd
import numpy as np
```

```python
numero_dias = 100
```

```python
numero_dias
```

```python
#mostra os 100 dias posteriores apartir de uma data
datas = pd.date_range(start='1/1/2021', periods=numero_dias)
```

```python
datas
```

```python
#cria dataframe com os valores das datas
df = pd.DataFrame(range(numero_dias), columns=['number'], index=datas)
```
