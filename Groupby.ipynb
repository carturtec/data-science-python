```python
import pandas as pd
import numpy as np
```

```python
#criando dataframe
data = {
        'Nome': ['jorge', 'carlos', 'roberta', 'ana', 'bruno', 'vera'],
        'Classe': ['júnior', 'júnior', 'pleno', 'pleno', 'sênior', 'sênior'],        
        'Venda': [200, 400, 300, 120, 140, 345]
       }
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
#agrupa os dados e exibe por um determinado parâmetro
group = df.groupby("Classe")
```

```python
#soma os grupos por classe mas pega os valores de vendas
group.sum()
```

```python
#tira a média de vendas
group.mean()
```

```python
#mostra quem vendeu menos
df.groupby("Classe").min()
```

```python
df.groupby("Classe").sum()
```

```python
df.groupby("Classe").mean()
```

```python
#mostra quem vendeu mais
df.groupby("Classe").max()
```
