```python
esquerda = pd.DataFrame(
                         {
                           'key': ['K0', 'K1', 'K2', 'K3'],
                           'A': ['A0', 'A1', 'A2', 'A3'],
                           'B': ['B0', 'B1', 'B2', 'B3']
                         })
   
direita = pd.DataFrame(
                        {
                           'key': ['K0', 'K1', 'K2', 'K3'],
                           'C': ['C0', 'C1', 'C2', 'C3'],
                           'D': ['D0', 'D1', 'D2', 'D3']
                        })    
```

```python
esquerda
```

```python
direita
```

```python
#mescla as duas dataframe e mantem a key ponto em comum
pd.merge(esquerda,direita, on='key')
```

```python
esquerda = pd.DataFrame(
                        {
                         'key1': ['K0', 'K0', 'K1', 'K2'],
                         'key2': ['K0', 'K1', 'K0', 'K1'],
                         'A': ['A0', 'A1', 'A2', 'A3'],
                         'B': ['B0', 'B1', 'B2', 'B3']
                        })
    
direita = pd.DataFrame(
                       {
                         'key1': ['K0', 'K1', 'K1', 'K2'],
                         'key2': ['K0', 'K0', 'K0', 'K0'],
                         'C': ['C0', 'C1', 'C2', 'C3'],
                         'D': ['D0', 'D1', 'D2', 'D3']
                       })
```

```python
esquerda
```

```python
direita
```

```python
#procura os indices em comum nas duas dataframes e junta
pd.merge(esquerda, direita, on=['key1', 'key2'])
```

```python
direita
```

```python
esquerda
```

```python
#insere todos os valores. observe que a sequencia k0 k1 só tem na 
#esquerda e no da direita ele só insere NaN pois não existe essa sequência k0 k1
pd.merge(esquerda, direita, how='outer', on=['key1', 'key2'])
```

```python
#isere de quaquer jeito os elementeos da direita
pd.merge(esquerda, direita, how='right', on=['key1', 'key2'])
```

```python
#isere de quaquer jeito os elementeos da esquerda
pd.merge(esquerda, direita, how='left', on=['key1', 'key2'])
```

```python
esquerda = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

direita = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
esquerda
```

```python
direita
```

```python
#o join puxa direto do dataframe e não de uma cópia como o pd pandas e anula os da direita como o k3
esquerda.join(direita)
```

```python
#o join puxa direto do dataframe e não de uma cópia como o pd pandas isere todos os dados dos dataframes
esquerda.join(direita, how='outer')
```

```python
#o join puxa direto do dataframe e não de uma cópia como o pd pandas isere os valores da esquerda
esquerda.join(direita, how='left')
```

```python
#o join puxa direto do dataframe e não de uma cópia como o pd pandas isere os valores da esquerda
esquerda.join(direita, how='right')
```
